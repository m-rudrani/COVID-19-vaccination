```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\rudrani\\OneDrive\\Documents\\college work\\dsbda miniproject\\covid_vaccine_statewise.csv")
```

```python
#a
# describe the dataset 
dataset_description = df.describe()
print(dataset_description)
```

```python
#b
print('\n')
# Number of persons state-wise vaccinated for the first dose in India
first_dose_counts = df.groupby('State')['First Dose Administered'].sum()
print(first_dose_counts)
```

```python
#c
print('\n')
# Number of persons state-wise vaccinated for the second dose in India
second_dose_counts = df.groupby('State')['Second Dose Administered'].sum()
print(second_dose_counts)

```

```python
#d
print('\n')
#Number of males vaccinated
males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print(males_vaccinated)
```

```python
#e
print('\n')
#Number of females vaccinated
females_vaccinated = df['Female(Individuals Vaccinated)'].sum()
print(females_vaccinated)
```

```python
# Bar plot for first dose vaccinated state-wise
plt.figure(figsize=(12, 6))
first_dose_counts.plot(kind='bar')
plt.title('Number of Persons Vaccinated for First Dose (State-wise)')
plt.xlabel('State')
plt.ylabel('Number of Persons')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for second dose vaccinated state-wise
plt.figure(figsize=(12, 6))
second_dose_counts.plot(kind='bar')
plt.title('Number of Persons Vaccinated for Second Dose (State-wise)')
plt.xlabel('State')
plt.ylabel('Number of Persons')
plt.xticks(rotation=90)
plt.show()
```

```python
# Data for the pie chart
labels = ['Males Vaccinated', 'Females Vaccinated']
sizes = [males_vaccinated, females_vaccinated]
# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
# Add a title
plt.title('Distribution of Vaccination by Gender')
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
# Display the chart
plt.show()
```
